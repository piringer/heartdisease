### Add new column BMI  - instead of height and weight
formula is from the CDC website
\begin{equation}
BMI = \frac{weight(kg)}{height(m)^2 }
\end{equation}

```python
# add BMI instead of height and weight 
# BMI = kg/m^2
# round to 1 decimal places

df['BMI'] = np.round((df.WEIGHT) / ((df.HEIGHT/100)**2), 1)
```

### Calculate the mean of the two systolic blood pressure columns and 
### do the same for the diastolic blood pressure columns

```python
# combine SYS1 and SYS2: use the average, round to 1 decimal places
# same for DIA1 and DIA2

df['SYS_mean'] = np.round(np.mean([df['SYS1'], df['SYS2']], axis=0), 1)
df['DIA_mean'] = np.round(np.mean([df['DIA1'], df['DIA2']], axis=0), 1)
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
    plt.legend(loc="center right", fontsize=16) 
    plt.xlabel("Threshold", fontsize=16)        
    plt.grid(True)                              
    #plt.axis([-5, 5, 0, 1])             

recall_90_precision = recalls[np.argmax(precisions >= 0.90)]
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]


plt.figure(figsize=(10,5))                                                                  
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
plt.plot([threshold_90_precision, threshold_90_precision], [0., 0.9], "r:")                 
plt.plot([-5, threshold_90_precision], [0.9, 0.9], "r:")                                
plt.plot([-5, threshold_90_precision], [recall_90_precision, recall_90_precision], "r:")
plt.plot([threshold_90_precision], [0.9], "ro")                                             
plt.plot([threshold_90_precision], [recall_90_precision], "ro")                                                                      # Not shown
plt.show()
```

```python
# split data into training and validation sets
X_valid, X_train_m = features_train[:450],  features_train[450:]
y_valid,  y_train_m = response_train[:450],  response_train[450:]
```

```python
# after playing with the hyperparameters, this model seemed reasonable:
# since the validation loss is very close to the training loss it is probably not overfitted
model = keras.models.Sequential([
    keras.layers.Dense(5, activation="relu", input_shape=[33]),
    keras.layers.Dense(3, activation="relu"),
    keras.layers.Dense(3, activation="softmax"), 
    keras.layers.Dense(2)
])
```

```python
# compile ANN model
model.compile(loss='mse', 
              optimizer=keras.optimizers.SGD(lr=1e-2),
              metrics=["accuracy"])
early_stopping_cb = keras.callbacks.EarlyStopping(patience=10,
                                                  restore_best_weights=True)

# fit model on the training data and use validation 
# use large number of epochs with early stopping
history = model.fit(X_train_m, y_train_m, epochs=400,
                    validation_data=(X_valid, y_valid),
                    callbacks=[early_stopping_cb])
```
